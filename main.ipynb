# Setup

```python
import pandas as pd
import numpy as np
import re
import datetime
import youtube_dl
import ast
from sklearn.model_selection import train_test_split
import requests as rq
import joblib as jb
import bs4 as bs4
import json
```

# Processing data

```python
queries = ["machine+learning", "data+science", "kaggle", "valorant", "investimentos"]
```

```python
ydl = youtube_dl.YoutubeDL({'ignoreerrors': True})
# resultados = []
for query in queries[3:]:
    r = ydl.extract_info("ytsearchdate2000:{}".format(query), download=False)
    for entry in r['entries']:
        if entry is not None:
            entry['query'] = query
    resultados += r['entries']
```

```python
resultados = [e for e in resultados if e != None]
```

```python
df = pd.DataFrame(resultados)
df['upload_date'] = pd.to_datetime(df["upload_date"])
df["days_since_upload"] = (pd.to_datetime("2021-12-03") - df["upload_date"])/np.timedelta64(1, 'D')
df.head()
```

```python
columns = ['id', 'title', 'upload_date', 'query', 'view_count', 'thumbnails', 'description', 'uploader']

df[columns].to_csv("csv/raw_data.csv", index=False)
```

# Split Train test

```python
df = pd.read_csv("csv/raw_data.csv")
train, test = train_test_split(df, stratify=df["query"],random_state=42,test_size=0.9)
```

```python
train["thumbnails"] = train["thumbnails"].apply(lambda y: ast.literal_eval(y)[0]['url'])
```

```python
train.to_csv("csv/train.csv", index=False)
test.to_csv("csv/test.csv", index=False)
```
